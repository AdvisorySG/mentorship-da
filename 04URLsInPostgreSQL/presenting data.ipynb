```python
import matplotlib.pyplot as plt
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

```

```python
industries = 'advertising', 'arts', 'banking', 'consultancy', 'data science' \
, 'infocom', 'venture capital and startups', 'biopharm', 'climate change' \
, 'design', 'energy and chemicals', 'healthcare', 'HR', 'legal' \
,'manufacturing', 'media', 'education', 'real estate', 'retail', 'security'\
, 'social impact & CSR', 'social service', 'training and adult education'\

sizes = [353, 
         424+71, 
         194+87+2933+66+52+193+13746+71+58, 
         87+2041+71+71+114, 
         425+58+66+52+1225+263,
         52+52+248+114+1981+263,
         193+71+567+855, 
         146+135,
         448, 
         364,
         113,
         2090+135,
         196+329,
         1199,
         66,
         1185+71,
         146,
         194,
         123,
         52,
         220,
         211,
         105]
```

```python
dictionary = {industries[i]:sizes[i] for i in range(len(sizes))}
temp = list(dictionary.items())
sort = sorted(temp, key=lambda x:x[1], reverse=True)
s = []
l = []
for i in sort:
    a,b = i
    l.append(a)
    s.append(b)
plt.figure(figsize=(20,5))
graph = plt.bar(l[:10], s[:10])
i = 0
total=sum(s)
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             f'{s[i]/total*100:.3f}'+'%',
             ha='center',
             weight='bold')
    i+=1
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(s, labels=l)
```
