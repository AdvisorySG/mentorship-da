```python
df = pd.read_csv('data/mentors.csv')
df['industries'] = df['industries'].apply(lambda x: literal_eval(x) if pd.notnull(x) else [])

mentors_info = {}
for index, row in df.iterrows():
    mentors_info[row['mentor_name']] = {
        'industries': row['industries'],
        'organisation': row['organisation']
    }
```

```python
mentors_info = mentors_info.keys()
```

```python
# simulate 2852 rows of 3 columns of mentor names mentor_1, mentor_2, mentor_3
# simulate from mentors_info list of names, each row has to have 3 distinct, unique names
# each mentor should have equal probability of being selected

# look at industries too

import pandas as pd
import matplotlib.pyplot as plt

import random

df_demo = pd.DataFrame(columns=['mentor_1', 'mentor_2', 'mentor_3'])

mentors_info = range(0, 205)

for i in range(376):
    mentors = random.sample(mentors_info, k=3)
    df_demo.loc[i] = mentors

# transpose all 3 columns into 1 column
df_demo = df_demo.melt(var_name='mentor', value_name='mentor_name')
# display(df_demo.head())

# popularity of mentors distribution
df_demo['mentor_name'].value_counts().hist(bins=range(0, max(df_demo['mentor_name'].value_counts()) + 6, 3), figsize=(10, 5))
# df_demo['mentor_name'].value_counts().hist(bins=range(0, max(df_demo['mentor_name'].value_counts()) + 6, 5), figsize=(10, 5))

# fix yaxis to 100
plt.ylim(0, 100)

# add x and y axis labels
# plt.xlabel('Number of applications')
plt.ylabel('Number of mentors')
plt.title('Distribution of popularity of mentors')

plt.show()
```
