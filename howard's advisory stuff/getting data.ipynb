```python
#gettin files
import os 
filesdir = os.path.dirname('C:/Users/howar/Downloads/Telegram Desktop/')
files = os.listdir(filesdir)
files
```

```python
#getting list of mentors

import pandas as pd
file = '2020w1.csv'
filedir = os.path.join(filesdir , file)
data = pd.read_csv(filedir)
print(data.isnull().sum())
data.dropna(how='all', inplace=True)
mentors = set()

for column in range(4):
    for mentor in data.iloc[:, column]:
        mentors.add(mentor)
        
mentors
```

```python
#finding frequency of mentors

frequency_dic = {x:0 for x in mentors}
print(type(frequency_dic))
for column in range(4):
    for mentor in data.iloc[:, column]:
        frequency_dic[mentor] += 1

sorted_data = pd.DataFrame(sorted(list(frequency_dic.items()),key=lambda x: x[1], reverse=True))
sorted_data
```

```python
#plotting frequency of mentors
import matplotlib.pyplot as plt

plt.bar(sorted_data.iloc[1:200,0], sorted_data.iloc[1:200,1])
```
